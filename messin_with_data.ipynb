```python
import json
import pandas as pd
from googletrans import Translator
import time
```

```python
with open('filtered_data.json', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
length = len(df)
length
```

```python
variety = df['ListTitle'].value_counts()
variety.head(20)
```

```python
df_1 = df.drop(df[(df.ListTitle == 'default list') | (df.ListTitle == 'to do') | (df.ListTitle == 'to do list') | (df.ListTitle == 'things to do') | (df.ListTitle == 'to do today')].index)
```

```python
df_1
```

```python
variety_1 = df_1['ListTitle'].value_counts()
variety_1.head(20)
```

```python
translator = Translator()

translated_data = []

count = 0

for index, row in df_1.iterrows():
    count += 1
    print(count)
    time.sleep(0.3)
    original_text = row['TaskTitle']
    translated_text = translator.translate(original_text, dest='ru')
    translated_data.append({"original": original_text, "translated": translated_text.text})
```

```python
translated_data
```

```python
with open('translated_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(translated_data, json_file, ensure_ascii=False, indent=4)
```
